```python
mainURL = 'https://www.farsnews.com'
def address_maker(cat=3,subcat=-1, year=1397,month=8,day=1,page=1):
    return(mainURL+'/archive?cat='+str(cat)+'&subcat='\
           +str(subcat)+'&date='+str(year)+'%2F'+str(month)+'%2F'+str(day)+'&p='+str(page))
```

```python
# txt = d1['0'].iloc[20]
def DataSetMaker(txt):

    cat,subcat,time,date,teller,title,summery,body = '','','','','','','',''
    html = BeautifulSoup(txt , 'html.parser')
#     print('      ',html)
    ######Cat
    top = html.find(class_='header py-2 mb-4 d-flex flex-wrap justify-content-between align-items-center text-right px-2 border-bottom')
#     print('      ',top)
    try:
        
        catField = top.findAll('div')[0].findAll('a')
#         print('      ',11)
        cat    = catField[0].get_text().replace('  ','').replace('\r\n','') #cat
#         print('      ',12)
        try:
            subcat = catField[1].get_text().replace('  ','').replace('\r\n','')
        except IndexError:
            subcat = 'dirty'
#          #subcat
    
#         print('      ',1)
    ######
    ######Time&Date
    
        dateField = top.findAll('div')[1]
        time = dateField.findAll('span')[0].get_text() #time
        date = dateField.findAll('span')[2].get_text() #date
#         print('      ',2)
    ###########

    ##########News
    
        header = html.find(class_='top')
        teller   = header.find(class_='headline d-block mb-1').get_text()#Teller
        title    = header.find(class_='title mb-2 d-block text-justify').get_text()#title
        summery  = header.find(class_='lead p-2 text-justify').get_text().replace('  ','')#Summery
#         imageUrl = header.find('img').get('src')#imageLinks
#         print('      ',3)
    ######body
    
        b = html.find(class_='nt-body text-right mt-4')
        body = b.get_text()#body

        return({
              'cat'     :cat,
              'subcat'  :subcat,
              'time'    :time,
              'date'    :date,
              'teller'  :teller,
              'title'   :title,
              'summery' :summery,
              'body'    :body
        })
    except:
        print('     ','whhyyy!!!!')
        return({
              'cat'     :cat,
              'subcat'  :subcat,
              'time'    :time,
              'date'    :date,
              'teller'  :teller,
              'title'   :title,
              'summery' :summery,
              'body'    :body
        })
```

```python
from multiprocessing import Pool

if __name__ == '__main__':
    p = Pool(8)
    tmp = list(p.imap(DataSetMaker,d1))
d1Done = pd.DataFrame(tmp)
d1Done
```
